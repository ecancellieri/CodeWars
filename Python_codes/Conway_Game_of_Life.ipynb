```python
# https://www.codewars.com/kata/52423db9add6f6fc39000354
# Conway's Game of Life - Unlimited Edition
import numpy as np

def n_live_neighbours(cells,i,j):
    idx_list = [[i-1,j-1], [i-1,j], [i-1,j+1], [i,j-1], [i,j+1], [i+1,j-1], [i+1,j], [i+1,j+1]]
    n = 0
    for idxs in idx_list:
        try:
            aux = cells[idxs[0],idxs[1]]
        except IndexError:
            aux = 0
        n = n + aux
    return n

def reduce_universe(cells):
    j = -1
    while j != 1:
        if np.sum(cells[0,:]) == 0:
            cells = cells[1:,:]
        else:
            j = 1
    j = -1
    while j != 1:
        if np.sum(cells[:,0]) == 0:
            cells = cells[:,1:]
        else:
            j = 1
    j = -1
    while j != 1:
        if np.sum(cells[:,-1]) == 0:
            cells = cells[:,0:-1]
        else:
            j = 1
    j = -1
    while j != 1:
        if np.sum(cells[-1,:]) == 0:
            cells = cells[0:-1,:]
        else:
            j = 1
    return np.array(cells)


def get_generation(cells, generations):
    import numpy as np
    cells = np.array(cells)
    for g in range(generations):
        nx = cells.shape[0]
        ny = cells.shape[1]
        aux_cells = np.zeros((nx+2,ny+2)).astype(int)
        for ix in range(nx):
            for iy in range(ny):
                aux_cells[ix+1,iy+1] = cells[ix,iy]
        cells = aux_cells
        aux_cells = np.zeros((nx+2,ny+2)).astype(int)
        for ix in range(nx+2):
            for iy in range(ny+2):
                n = n_live_neighbours(cells,ix,iy)
                if cells[ix,iy] == 0:
                    if n == 3:
                        aux_cells[ix,iy] = 1
                else:
                    if n not in (2,3):
                        aux_cells[ix,iy] = 0
                    else:
                        aux_cells[ix,iy] = cells[ix,iy]
        cells = reduce_universe(aux_cells)
    if np.sum(cells) == 0:
        cells = [[]]
    else:
        cells = cells.tolist()
    return cells
```
